```python
import torchvision.io as tv_io
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
def detect_and_plot_edges_hybrid(image_path):
    try:
        # Read image with torchvision (CxHxW tensor)
        image_tensor = tv_io.read_image(image_path)
    except RuntimeError:
        print("Error loading image")
        return

    # Convert to NumPy array and prepare for OpenCV (HxWxC BGR format)
    image_np = image_tensor.numpy().transpose(1, 2, 0)  # CHW -> HWC
    image_np = cv2.cvtColor(image_np, cv2.COLOR_RGB2BGR)  # Torch RGB to OpenCV BGR

    # Convert to grayscale and detect edges
    gray = cv2.cvtColor(image_np, cv2.COLOR_BGR2GRAY)
    
    edges = cv2.Canny(gray, 50, 150)

    # Prepare original image for matplotlib (RGB format)
    original_for_display = cv2.cvtColor(image_np, cv2.COLOR_BGR2RGB)

    # Plotting
    plt.figure(figsize=(10, 5))
    
    plt.subplot(1, 2, 1)
    plt.imshow(original_for_display)
    plt.title('Original Image')
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    plt.imshow(edges, cmap='gray')
    plt.title('Detected Edges')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
def sobel_edge_detection(image_path):
    # Read the image
    img = cv2.imread(image_path)
    if img is None:
        print("Error: Image not found or invalid path")
        return
    
    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Apply Gaussian blur to reduce noise
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    
    # Sobel Edge Detection
    sobel_x = cv2.Sobel(blurred, cv2.CV_64F, 1, 0, ksize=3)  # Horizontal edges
    sobel_y = cv2.Sobel(blurred, cv2.CV_64F, 0, 1, ksize=3)  # Vertical edges
    
    # Convert to absolute values and combine
    abs_sobel_x = cv2.convertScaleAbs(sobel_x)
    abs_sobel_y = cv2.convertScaleAbs(sobel_y)
    
    # Combine both directional edges
    combined = cv2.addWeighted(abs_sobel_x, 0.5, abs_sobel_y, 0.5, 0)
    
    # Create plots
    plt.figure(figsize=(15, 6))
    
    # Original image
    plt.subplot(1, 3, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')
    plt.axis('off')
    
    # Horizontal edges
    plt.subplot(1, 3, 2)
    plt.imshow(abs_sobel_x, cmap='gray')
    plt.title('Sobel X Direction')
    plt.axis('off')
    
    # Combined edges
    plt.subplot(1, 3, 3)
    plt.imshow(combined, cmap='gray')
    plt.title('Combined Sobel Edges')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()
```
